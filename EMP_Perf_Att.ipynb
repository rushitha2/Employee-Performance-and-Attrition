#### Handling two gender columns

```python
unique_values = df['gender'].unique()
print(unique_values)
```

```python
v=np.array(['Male','Female'])
```

```python
unique_values = np.setdiff1d(unique_values,v)
print(unique_values)
```

```python
df['new_gender']= df.apply(lambda row: row['Gender'] if row['gender'] in unique_values else row['gender'],axis=1)
```

```python
df[['gender','Gender','new_gender']]
```

```python
df['new_gender'].unique()
```

```python
df['new_gender'].value_counts()
```

```python
df.drop(columns=['gender', 'Gender'], inplace=True)
df
```

```python
df.rename(columns={'new_gender': 'gender'}, inplace=True)
df
```

#### data cleaning

```python
df
```
